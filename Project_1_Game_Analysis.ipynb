```python
#Task 11

pivot_table = df.pivot_table(
    index=['category', 'type'], # index: define rows(grouped as these rows)
    values=['price', 'rating', 'reviews'], #the columns which they are used for agg function
    aggfunc='mean' #aggregation function which that used for mean, median, sum or ext
).round(2)

pivot_table = pivot_table.rename(columns={
    'price': 'mean_price',
    'rating': 'mean_rating',
    'reviews' : 'mean_reviews'
})

pivot_table.to_csv('pivot_table.csv', sep=',')
```

```python
#Task 12

def delete_size_column(value):
  if isinstance(value, str):
    value = value.strip().upper()
    if value.endswith('M'):
      value = value.replace('M', '')
      value = float(value)
      value = value * 1024
      return value
    elif value.endswith('K'):
      value = value.replace('K', '')
      value = float(value)
      return value


df['size'] = df['size'].apply(delete_size_column)

```
